```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('classic')
data = sns.load_dataset('flights')
data.head()
```

```python
x=data.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,annot=True,linewidths=0.5,cmap='summer',fmt='d')
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,z_score=0,annot=True,metric='correlation')
```
