```python
import matplotlib.pyplot as plt
import numpy as np

ages_x = [25,26,27,28,29,30]
dev_y = [625,676,729,812,871,900]
py_dev_y = [650,725,835,890,900,960]
```

```python
print(plt.style.available)
```

```python
plt.style.use('classic')
plt.plot(ages_x,dev_y,color='k',linestyle='--',marker='o',label='all devs')
plt.bar(ages_x,py_dev_y,color='#5a7d9a',label='python devs')
plt.xlabel('Ages')
plt.ylabel('wages/day')
plt.title('median wages by age')
plt.legend()
plt.grid()
plt.show()
```

```python
x_indexes = np.arange(len(ages_x))
width = 0.25
```

```python
plt.style.use('classic')
plt.bar(x_indexes - width,dev_y,width = width,color='k',label='all devs')
plt.bar(x_indexes,py_dev_y,width = width,color='#5a7d9a',label='python devs')
plt.xlabel('Ages')
plt.ylabel('wages/day')
plt.title('median wages by age')
plt.xticks(ticks=x_indexes,labels=ages_x)
plt.legend(loc='upper left')
plt.tight_layout()
plt.grid()
plt.show()
```
