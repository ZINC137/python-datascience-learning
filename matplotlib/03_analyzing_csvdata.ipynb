```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv('datas/data.csv')
```

```python
df.head()
```

```python
JavaScript = df.LanguagesWorkedWith.str.contains('JavaScript', na=False).sum()
JavaScript
```

```python
Python = df['LanguagesWorkedWith'].str.contains('Python', na=False).sum()

Java = df['LanguagesWorkedWith'].str.contains(r'\bJava\b', regex=True, na=False).sum()

Cpp = df['LanguagesWorkedWith'].str.contains('C++', regex=False, na=False).sum()
```

```python
languages = ['JavaScript', 'Python', 'Java', 'C++']

users = [JavaScript, Python, Java, Cpp]

plt.style.use('ggplot')

plt.barh(languages, users, height=0.4)

plt.xlabel('Number of People Who Use')
plt.ylabel('Languages')
plt.title('Most Popular Languages')

plt.xticks([0, 20000, 40000, 60000])

plt.show()
```
