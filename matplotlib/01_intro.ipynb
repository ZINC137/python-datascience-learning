```python
dev_x = [25,26,27,28,29,30]
dev_y = [625,676,729,812,871,900]
```

```python
py_dev_y = [650,725,835,890,900,960]
```

```python
print(plt.style.available)
```

```python
plt.style.use('classic')
plt.plot(dev_x,dev_y,color='k',linestyle='--',marker='o',label='all devs')
plt.plot(dev_x,py_dev_y,color='#5a7d9a',label='python devs')
plt.xlabel('Ages')
plt.ylabel('wages/day')
plt.title('median wages by age')
plt.legend()
plt.grid()
plt.savefig('plot.png')
plt.show()
```
